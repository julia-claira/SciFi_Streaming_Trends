```python
#Import Dependencies

import pandas as pd
import math
import matplotlib.pyplot as plt
import sqlite3
import numpy as np
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
#Load SQLite DB
connection = sqlite3.connect('IMDB.db')
conn=connection.cursor()
```

*****************************
Explore Data & Graph
*****************************

```python
#See which genres get the highest average rating on IMDB

#Grab Main Genre Averages
scifi=pd.read_sql('select sum(averageRating) as mean from titles where genres like "%Sci-Fi%"',connection)['mean'][0]
horror=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Horror%"',connection)['mean'][0]
adventure=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Adventure%"',connection)['mean'][0]
drama=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Drama%"',connection)['mean'][0]
fantasy=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Fantasy%"',connection)['mean'][0]
action=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Action%"',connection)['mean'][0]
comedy=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Come%"',connection)['mean'][0]
documentary=pd.read_sql('select avg(averageRating) as mean from titles where genres like "%Docu%"',connection)['mean'][0]
```

```python
#Plot Genre Rating Average

#Create Dictionary to Store Values
genre_avg_Y=[scifi,horror,adventure,drama, \
          fantasy,action,comedy,documentary]
genre_avg_X=np.arange(len(genre_avg_Y))
labels=['Sci-Fi','Horror','Advent','Drama', \
          'Fantasy','Action','Comedy','Doc']

#Plot Category pie Chart
fig2=plt.figure(figsize=(10,8))
plt.xticks(np.arange(8), labels, fontsize=18)
plt.yticks(fontsize=14)

#Layout
plt.title('Genre Rating Average', fontsize=20,color='r')

#Plot Graph
plt.bar(genre_avg_X, genre_avg_Y)
plt.tight_layout()


#Show
plt.show()
```

```python
#See which genres get the highest average votes on IMDB

#Grab Main Votes Averages
scifi=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Sci-Fi%"',connection)['mean'][0]
horror=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Horror%"',connection)['mean'][0]
adventure=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Adventure%"',connection)['mean'][0]
drama=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Drama%"',connection)['mean'][0]
fantasy=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Fantasy%"',connection)['mean'][0]
action=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Action%"',connection)['mean'][0]
comedy=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Comedy%"',connection)['mean'][0]
documentary=pd.read_sql('select avg(numVotes) as mean from titles where genres like "%Docu%"',connection)['mean'][0]
```

```python
#Plot Genre Average Votes

#Create Dictionary to Store Values
genre_avg_Y=[scifi,horror,adventure,drama, \
          fantasy,action,comedy,documentary]
genre_avg_X=np.arange(len(genre_avg_Y))
labels=['Sci-Fi','Horror','Advent','Drama', \
          'Fantasy','Action','Comedy','Doc']

#Plot Category pie Chart
fig2=plt.figure(figsize=(10,8))
plt.yticks(np.arange(8), labels, fontsize=18)
plt.xticks(fontsize=14)

#Layout
plt.title('Genre Rating Votes', fontsize=20,color='b')

#Plot Graph
plt.barh(genre_avg_X, genre_avg_Y,color='orange')
plt.tight_layout()


#Show
plt.show()
```

```python
men=pd.read_sql('select view as mean from flix where genre like "%Science%" and gender="men"',connection)['mean']
women=pd.read_sql('select view as mean from flix where genre like "%Science%" and gender="women"',connection)['mean']
```

```python
#Compare the mean sci-fi views for men and women

men_views=men.str.split(',')
men_corr=[]
for view in men_views:
    men_corr.append(int(view[0]))
    
women_views=women.str.split(',')
women_corr=[]
for view in women_views:
    women_corr.append(int(view[0]))
```

```python
print(f' p-value: {ttest_ind(men_corr, women_corr).pvalue}')
```
